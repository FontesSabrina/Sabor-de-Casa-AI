## Pipeline ETL com IA Generativa

Projeto desenvolvido para o desafio DIO - Explorando IA Generativa em um Pipeline de ETL com Python.

Objetivo:
Personalizar mensagens para clientes de uma FoodTech brasileira utilizando IA.

```python
# ==========================================
# 🚀 PROJETO: PIPELINE ETL COM IA (SABOR DE CASA)
# ==========================================

# 1. IMPORTS (Apenas o necessário, sem repetições)
import pandas as pd
from openai import OpenAI
import time

# --- ETAPA 1: EXTRACT (Extração) ---
# Lendo os dados que você já conferiu e estão certinhos
arquivo = 'clientes_sabor_de_casa_ai.xlsx'
df = pd.read_excel(arquivo)

print("=" * 40)
print("       TABELA DE CLIENTES (EXTRACT)")
print("=" * 40)
display(df.head()) # Mostra a tabela inicial


# --- ETAPA 2: TRANSFORM (Transformação com IA) ---
# Usando a sua chave que configuramos
client = OpenAI(api_key="****************************")

def gerar_dados_ia(linha):
    nome = linha['nome']
    prato = linha['prato_favorito']

    print(f"✍️ Criando conteúdo para: {nome}...")

    # Prompt otimizado para a IA não esquecer os separadores
    prompt = f"Crie uma mensagem curta para {nome} que ama {prato}. Responda obrigatoriamente no formato: Mensagem # Oferta # Dica"

    try:
        response = client.chat.completions.create(
            model="gpt-3.5-turbo",
            messages=[{"role": "user", "content": prompt}]
        )
        # Divide a resposta da IA em 3 partes pelo símbolo #
        res = response.choices[0].message.content.split('#')

        # Se a IA enviou as 3 partes certinho, limpamos os espaços
        if len(res) >= 3:
            return pd.Series([res[0].strip(), res[1].strip(), res[2].strip()])
        else:
            return pd.Series([res[0].strip(), "Oferta Especial no App!", "Experimente hoje!"])

    except Exception as e:
        # Se houver erro de chave ou limite, preenche com erro amigável
        return pd.Series(["Erro na IA", "Erro na IA", "Erro na IA"])

# Aplicando a transformação para criar as colunas novas
df[['mensagem_personalizada', 'oferta', 'recomendacao']] = df.apply(gerar_dados_ia, axis=1)

print("\n✅ TRANSFORM CONCLUÍDO!")


# --- ETAPA 3: LOAD (Carregamento / Salvamento) ---
# Salvando o resultado final em Excel
df.to_excel('resultado_final_sabor_de_casa.xlsx', index=False)
print("💾 ARQUIVO 'resultado_final_sabor_de_casa.xlsx' SALVO!")


# --- VISUALIZAÇÃO FINAL (ESTILO DIO) ---
print("\n" + "=" * 40)
print("       MENSAGENS GERADAS (PREVIEW)")
print("=" * 40)

# Mostra o resultado de cada cliente na tela
for index, linha in df.iterrows():
    print(f"👤 Cliente: {linha['nome']}")
    print(f"📩 Mensagem: {linha['mensagem_personalizada']}")
    print(f"🎁 Oferta: {linha['oferta']}")
    print(f"💡 Dica: {linha['recomendacao']}")
    print("-" * 40)
```
